```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_473760.csv', skiprows=4)
country_metadata = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_473760.csv')
indicator_metadata = pd.read_csv('Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_473760.csv')

# Filter the data for the year 2022
population_2022 = population_data[['Country Name', '2022']].dropna()

# Sort the data to make the chart more readable
population_2022_sorted = population_2022.sort_values(by='2022', ascending=False).head(30)  # Top 30 countries for visibility

# Plot the data
plt.figure(figsize=(12, 10))
bars = plt.barh(population_2022_sorted['Country Name'], population_2022_sorted['2022'], color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Population (millions)')
plt.ylabel('Country')
plt.title('Top 30 Countries by Population in 2022')

# Add data labels to the bars
for bar in bars:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2, '{:,.0f}'.format(bar.get_width()),
             va='center', ha='left', fontsize=10)

# Invert y-axis for better readability
plt.gca().invert_yaxis()

# Customize grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout to prevent labels from being cut off
plt.tight_layout()

# Display the plot
plt.show()
```
